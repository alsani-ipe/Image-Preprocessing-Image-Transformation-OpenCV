```python
#import necessary libraries 

import matplotlib.pyplot as plt
import cv2
import numpy as np 
import warnings 
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
def image_Display(imageTitle, image):
    imageVariable= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize= (4,2))
    plt.imshow(imageVariable)
    plt.title(imageTitle)
    plt.show()
```

```python
image=cv2.imread("../input/catimage/cat.jpg")
image_Display("Display image",image)
```

# 3.1 𝐈𝐦𝐚𝐠𝐞 𝐓𝐫𝐚𝐧𝐬𝐥𝐚𝐭𝐢𝐨𝐧

**Image translation** is a basic image transformation operation that involves shifting an image's pixels horizontally and/or vertically. It's commonly used in image processing for tasks like image **registration**, **object tracking**, and **data augmentation**. Here's how you can perform image translation using Python with the OpenCV library:

```python
#Making a defined function for image translation

def imageTranslation(image, translationX, translationY):
    #translationX=input()
    #translationY=input()
    modifiedImage= np.float32([[1,0,translationX],
                             [0,1,translationY]])
    imageShift= cv2.warpAffine(image,modifiedImage,(image.shape[1],image.shape[0]))
    image_Display("Orginal Image",image)
    image_Display("Shifted Image",imageShift)
```

```python
imageTranslation(image,-150,100)
```

# 3.2 𝐈𝐦𝐚𝐠𝐞 𝐑𝐨𝐭𝐚𝐭𝐢𝐨𝐧

**Rotation** of an Image is one of the Image transform operations that can be applied on an Image. By Image rotation, the image is rotated about its center by specified number of degrees. Image rotation is a specialization of affine transformation.

```python
#Making a defined Function for image Rotation

def imageRoation(image,angle, scale):
    imageCenter=(image.shape[1]//2, image.shape[0]//2)
    imageRotate= cv2.getRotationMatrix2D(imageCenter, angle, scale)
    finalRotate=cv2.warpAffine(image, imageRotate,(image.shape[1],image.shape[0]))
    image_Display("Orginal Image", image)
    image_Display("Rotated Image", finalRotate)
```

```python
imageRoation(image,120,1)
```

# 3.3 𝐈𝐦𝐚𝐠𝐞 𝐑𝐞𝐬𝐢𝐳𝐢𝐧𝐠

Image resizing is necessary when you need to increase or decrease the total number of pixels, whereas remapping can occur when you are correcting for lens distortion or rotating an image. Zooming refers to increase the quantity of pixels, so that when you zoom an image.

```python
def imageResizing(image,imageSize):
    imageResize= cv2.resize(image,(imageSize,imageSize),interpolation=cv2.INTER_AREA)
    print("Before Resize image shape is: {}".format(image.shape))
    image_Display("Before Resize",image)
    print("After Resize image shape is: {}".format(imageResize.shape))
    image_Display("After Resize",imageResize)
```

```python
imageResizing(image,150)
```

# 3.4 𝐈𝐦𝐚𝐠𝐞 𝐅𝐥𝐢𝐩𝐩𝐢𝐧𝐠


Flipping an image in image processing involves reversing the pixel values along one or both axes, typically horizontally (left to right) or vertically (upside down). You can perform image flipping in Python using libraries like Pillow (PIL Fork) or OpenCV.

```python
def imageFlipping(image):
    imageFlip= cv2.flip(image,0)
    print("Before Flipping Image")
    image_Display("Orginal Display",image)
    print("After Flipping Image")
    image_Display("Flipped Image",imageFlip)
    
```

```python
imageFlipping(image)
```

# 3.5 𝐈𝐦𝐚𝐠𝐞 𝐂𝐫𝐨𝐩𝐩𝐢𝐧𝐠

Cropping an image in image processing involves selecting a specific region of interest (ROI) from the original image and discarding the rest. You can crop an image in Python using libraries like Pillow (PIL Fork) or OpenCV.

```python
def imageCropping(image):
    imageCrop= image[50:200, 50:200]
    print("Before Image Crop")
    image_Display("Orginal Image",image)
    print("After Image Crop")
    image_Display("Cropped Image",imageCrop)
```

```python
imageCropping(image)
```
